## Spatial block split vs. naive random split

```python
train_idx, val_idx, test_idx = spatial_block_split(config.EUROSAT_DIR)
print(f'block split -> train {len(train_idx)} / val {len(val_idx)} / test {len(test_idx)}')

train_idx_r, val_idx_r, test_idx_r = random_split(config.EUROSAT_DIR)
print(f'random split -> train {len(train_idx_r)} / val {len(val_idx_r)} / test {len(test_idx_r)}')
```
