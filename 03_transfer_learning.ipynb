# 03 — Transfer Learning (Module 1)

Two-phase fine-tuning: frozen backbone for 3 epochs, then unfreeze the last 2 conv blocks and drop LR 10x for 5 more epochs.

```python
import sys, subprocess
sys.path.append('..')
result = subprocess.run(
    ['python', '-m', 'src.train_transfer', '--backbone', 'resnet18',
     '--phase1-epochs', '3', '--phase2-epochs', '5'],
    cwd='..', capture_output=True, text=True
)
print(result.stdout)
print(result.stderr)
```

## Frozen vs. unfrozen comparison

```python
import json
from src import config
with open(config.REPORT_DIR / 'transfer_resnet18_metrics.json') as f:
    metrics = json.load(f)
print('frozen (phase 1) val acc: ', metrics['frozen_val_acc'])
print('unfrozen (phase 2) val acc:', metrics['unfrozen_val_acc'])
print('macro-F1:', metrics['macro_f1'])
```
